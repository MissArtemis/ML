```python
import seaborn as sns
Iris = sns.load_dataset("iris")
print(Iris)
```

```python
sns.pairplot(Iris,hue="species",size=2,diag_kind="hist",markers=["o","s","D"])
```

```python
from pandas.plotting import parallel_coordinates
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
parallel_coordinates(Iris,"species",alpha=0.8)
plt.title("Parallel Plot")
plt.show()
```

```python
import numpy as np
import pandas as pd
newdata = Iris.drop(["species"],axis=1)
datacor = np.corrcoef(newdata,rowvar=0)
datacor = pd.DataFrame(data=datacor,columns=newdata.columns,index=newdata.columns)
print(datacor)
```

```python
plt.figure(figsize=(6,6))
ax = sns.heatmap(data=datacor,square=True,annot=True,fmt=".3f",linewidths=.5,cbar_kws={"fraction":0.046,"pad":0.03})
ax.set_title("Iris Correaltion")
plt.show()
```
