```python
import matplotlib.pyplot as plt

#show plot
%matplotlib inline
#show the type of img
%config InlineBackend.figure_format="retina"
```

```python
from matplotlib.font_manager import FontProperties
# fonts = FontProperties(fname="/Lib/Font/微软雅黑.ttf",size=14)
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

```

```python
import numpy as np
X = np.linspace(1,15)
Y = np.sin(X)
plt.figure(figsize=(6,4))
plt.plot(X,Y,"r-o")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y = sin(x)")
plt.grid("on")
plt.show()

```

```python
xtick = [0,5,10,15]
xticklabel=["here is:"+str(x) for x in xtick]
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.plot(X,Y,"b-.s")
plt.xlabel(r"$\alpha$")
plt.ylabel(r"$\beta$")
plt.title("$y = \sum sin(x)$")
plt.subplot(2,2,2)
plt.hist(Y,20)
plt.text(0,7,"Hist")
plt.xlabel("Freq")
plt.ylabel("Num")
plt.title("Hist")
plt.subplot(2,1,2)
plt.step(X,Y,c="R",label="sin(x)",linewidth=3)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bar")
plt.legend(loc="Lower Right")
plt.xticks(xtick,xticklabel,rotation=45)
plt.subplots_adjust(hspace=0.35)
plt.show()
```

```python
#####################3D Plot############################
x = np.linspace(-4,4,num=50)
y = np.linspace(-4,4,num=50)
X,Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2+Y**2))
fig = plt.figure(figsize=(12,5))
```

```python
ax1 = fig.add_subplot(121,projection="3d")
```

```python
ax1.plot_surface(X,Y,Z,rstride=1,cstride=1,alpha=0.5,cmap=plt.cm.coolwarm)
```

```python
cset = ax1.contour(X,Y,Z,zdir="z",offset=1,cmap=cm.CMRmap)
ax1.set_xlabel("X")
ax1.set_xlim(-4,4)
ax1.set_ylabel("Y")
ax1.set_ylim(-4,4)
ax1.set_zlabel("Z")
ax1.set_zlim(-4,4)
ax1.set_title("surface and contour")
```

```python
theta = np.linspace(-4*np.pi,4*np.pi,100)
z = np.linspace(-2,2,100)
r = z**2+1
x = r*np.sin(theta)
y = r*np.cos(theta)
ax2 = plt.subplot(122,projection="3d")
ax2.plot(x,y,z)
ax2.scatter3D(x,y,z,c="r")
ax2.view_init(elev=20,azim=25)
ax2.set_title("Curve and Scatter")
plt.show()
```
